# matlpotlib.pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel("number")
plt.xlabel("indices")
plt.title("my first")
plt.grid()
plt.show()
```

```python
t=[1,2,3,4,5]
t2=[1,4,9,16,25]
plt.plot(t,t2,'r--',linewidth=5.0)
plt.ylabel("squar")
plt.xlabel("number")
plt.grid()
```

```python
t=[1,2,3,4,5, ]
t2=[1,4,9,16,25]
plt.plot(t,t2,'ro',label="numbers")
plt.ylabel("squar")
plt.xlabel("number")
plt.legend()
plt.grid()
```

```python
import numpy as np
t=np.arange(1,10,0.5)
plt.plot(t,t**2,"g-",label="^2")
plt.plot(t,t**2.2,"r^",label="^2.2")
plt.plot(t,t**2.4,"mo",label="^2.4")
plt.ylabel("multiple")
plt.xlabel("numbers")
plt.title("2nd graph")
plt.legend()
plt.show()
```

```python
x=np.arange(1,10,2)
y=2*x
x1=np.arange(1,10,2)
y1=3*x1

lines=plt.plot(x,y,x1,y1)

plt.setp(lines[0],color="r",linestyle="--",linewidth="2.5",label="square")
plt.setp(lines[1],color="g",linestyle="dotted",linewidth="5.5",label="cube")
plt.legend()
plt.ylabel("square and cubes")
plt.xlabel("numbers")
plt.title("Two line same time")
plt.grid()
```
