```python
plt.plot(seto.SepalLengthCm,seto.SepalWidthCm,"ro")
plt.plot(vir.SepalLengthCm,vir.SepalWidthCm,"mo")
plt.plot(ver.SepalLengthCm,ver.SepalWidthCm,"go")
plt.xlabel("Length of species")
plt.ylabel("Width of species")
plt.title("Sepal Length & Sepal width Graph")
plt.legend(['setosa','virginica','versicolor'])
plt.show()
```
