```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title("testv plot")
plt.xlabel("x")
plt.ylabel("y & z")
plt.legend(["y","z"])
plt.show()
```

```python
data2=pd.read_csv("countries.csv")
data2
```

```python
data2.country
```

```python
us_tf=data2.country=='United States'
us_tf
```

```python
us=data2[data2.country=='United States']
us
```

```python
china=data2[data2.country=='China']
china
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.xlabel("year")
plt.ylabel("Population in Million")
plt.legend(['United States','China'])
```

```python
afg=data2[data2.country=='Afghanistan']
afg
```

```python
plt.plot(us.year,us.population/us.population.iloc[0]*100)
plt.plot(china.year,china.population/china.population.iloc[0]*100)
plt.plot(afg.year,afg.population/afg.population.iloc[0]*100,"g--")
plt.xlabel("year")
plt.ylabel("Population increase in %")
plt.legend(['United States','China','Afghanistan'])
```
